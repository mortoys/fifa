```python
import pandas as pd
```

```python
exp = pd.read_csv('expectation.csv')
```

```python
dd = pd.read_csv('fifa2022.csv')
```

```python
data = pd.merge(dd, exp, on='xeid')
data = data[data['usable']]
```

```python
data = data.sort_values('date').reset_index(drop=True)
```

```python
data
```

```python
(data['h11e'] / data['h21e']).plot()
(data['h12e'] / data['h22e']).plot()
```

```python
(data['t1e'] / (data['h11e'] + data['h21e'])).plot()
(data['t2e'] / (data['h12e'] + data['h22e'])).plot()
```

```python
data['t1e'].mean() - data['t2e'].mean()
```

```python
(data['t1e'] - 0.276).rolling(200).mean().plot()
(data['t2e']).rolling(200).mean().plot()
```

```python
(data['t1e'] + data['t2e']).rolling(60).mean().plot()
```
